```python
#reading tweet_json.txt line by line into a DataFrame 
tweet_list = []
with open('tweet_json.txt', 'r') as file:
    for line in file:
        try:
            tweet = json.loads(line)
            tweet_list.append(tweet)
        except:
            continue
df_tweet = pd.DataFrame()
df_tweet['tweet_id'] = list(map(lambda tweet: tweet['id'],tweet_list))
df_tweet['favorite_count'] = list(map(lambda tweet: tweet['favorite_count'], tweet_list))
df_tweet['retweet_count'] = list(map(lambda tweet: tweet['retweet_count'], tweet_list))
df_tweet['created_at'] = list(map(lambda tweet: tweet['created_at'], tweet_list))
df_tweet['full_text'] = list(map(lambda tweet: tweet['full_text'], tweet_list))
```

###  Archive tweets datasets

###  Issue #1: Archive tweets datasets contains columns not in use for the analysis

#### Define: keep only original tweets and Drop unnecessary columns in archive tweets dataset eg retweets ratings

```python
# drop retweets records and keep onle original tweets
archiveClean = archiveClean[archiveClean['retweeted_status_id'].isnull()]
archiveClean.shape
```

#### Code

```python
#drop unnecessary columns
archiveClean.drop(columns =['in_reply_to_status_id','in_reply_to_user_id','retweeted_status_user_id','retweeted_status_id','retweeted_status_timestamp'], inplace = True)
```

###  Issue #2: Merging of the dog stage columns doggo, puppo, pupper and floofer

#### Define: Merge dog columns doggo, puppo, pupper and floofer to one column


#### Code

```python
cols = ['doggo','floofer','pupper', 'puppo']
archiveClean[cols] = archiveClean[cols].replace({'None':np.nan})
archiveClean.head()
```

```python
#unpivot 'doggo','floofer','pupper', 'puppo' and assign to a dog_nicknames dataframe
dog_nicknames = pd.melt(archiveClean, id_vars = ['tweet_id'], value_vars= ['doggo','floofer','pupper', 'puppo'],
       var_name = 'dog_name', value_name = 'value_name')
dog_nicknames.info()
```

```python
dog_nicknames.value_name.value_counts()
```

```python
#convert value column created to str type and replace nan with np.nan
dog_nicknames.value_name = dog_nicknames.value_name.astype(str)
dog_nicknames.value_name = dog_nicknames.value_name.replace({'nan': np.nan})
dog_nicknames.value_name.value_counts()
```

```python
#drop nan in column value_name 
dog_nicknames.dropna(subset = ['value_name'], inplace = True)
```

```python
dog_nicknames.info()
```

```python
#drop the unpivoted columns
archiveClean.drop(columns= ['doggo','floofer','pupper', 'puppo'], inplace = True)
```

```python
#join the 2 dataframes on tweet_id keeping all records from archiveClean
archiveClean = archiveClean.merge(dog_nicknames, how = 'left', on = 'tweet_id')
archiveClean
```

#### Test

```python
#set default figure size
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10, 4)
```

```python
dogname_value = df_Tweeter_clean.dog_name.value_counts().values
dogname = df_Tweeter_clean.dog_name.value_counts().index

plt.figure(figsize = (8,4))
plt.bar(dogname, height = dogname_value, width = 0.8)
plt.title('Number of dogs per dog type')
plt.xlabel('Dog Type');
```

```python
ratings = df_Tweeter_clean.groupby(['rating_numerator']).size().reset_index().rename(columns = {0: 'Rating Count'})
ratings
```

```python
#Distribution of the rating 
sns.barplot(x = 'rating_numerator', y='Rating Count' , data = ratings, color = 'blue')
plt.title('Rating Distribution')
plt.ylabel('No of occurrence');
```

```python
temp_ratings = df_Tweeter_clean.groupby(['dog_name'])[['rating_numerator']].mean().reset_index()
temp_ratings
```

```python
temp_retweets = df_Tweeter_clean.groupby(['dog_name'])[['retweet_count']].sum().reset_index()
plt.figure(figsize = (6,6))
plt.pie(temp_retweets.retweet_count, labels = temp_retweets.dog_name, autopct = '%.2f%%')
plt.title('Dog Type comparison in terms of retweet count');
```
